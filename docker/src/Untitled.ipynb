```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.ticker as mticker
import matplotlib.path as mpath
import numpy as np

# 50m解像度用の陸地データ作成
land_50m = cfeature.NaturalEarthFeature('physical', 'land', '50m',
                                        edgecolor='face',
                                        facecolor=cfeature.COLORS['land'])
```

```python
lons = np.tile(np.arange(110, 170, 0.5), (100, 1)) * np.pi * 2 / 180
lats = np.tile(np.arange(10, 60, 0.5), (120, 1)).T * np.pi * 2 / 180

data = np.tile(np.arange(0, 12, 0.1), (100, 1))

proj = ccrs.LambertConformal(central_longitude=140.0,
                                               central_latitude=30,
                                               standard_parallels=(30,60))
ax = plt.axes(projection=proj)
ax.set_extent([110,170,10,60],ccrs.PlateCarree())
ax.coastlines(resolution='50m', lw=0.5)
ax.gridlines(xlocs=mticker.MultipleLocator(10), 
             ylocs=mticker.MultipleLocator(10), 
             linestyle='-', 
             color='gray')
ax.contourf(lons, lats, data, 60,transform=proj)
plt.show()
```
